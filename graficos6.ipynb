```python
#preguntar a 5 estudiantes su peso despues sacar la media y la desbiacion estandar (representarlo en un diagrama de sector)
```

```python
import numpy as np
import matplotlib.pyplot as plt

weights = []
for i in range(5):
  while True:
    try:
      weight = float(input(f"Ingrese el peso del estudiante {i+1}: "))
      weights.append(weight)
      break
    except ValueError:
      print("Entrada inválida. Por favor, ingrese un número para el peso.")

print("Pesos ingresados:", weights)


mean_weight = np.mean(weights)
std_deviation_weight = np.std(weights)

print(f"La media del peso es: {mean_weight:.2f}")
print(f"La desviación estándar del peso es: {std_deviation_weight:.2f}")


students = [f"Estudiante {i+1}" for i in range(5)]

fig, ax = plt.subplots()
ax.pie(weights, labels=students, autopct='%1.1f%%')
ax.set_title('Peso de 5 Estudiantes')
plt.show()

labels = ['Media', 'Desviación Estándar']
values = [mean_weight, std_deviation_weight]

fig, ax = plt.subplots()
ax.bar(labels, values, color=['blue', 'orange'])
ax.set_ylabel('Peso')
ax.set_title('Media y Desviación Estándar del Peso de los Estudiantes')

plt.show()
```

```python
# ingresar y calcular el porcentaje de cuantos varones y mujeres hay
# representar en un diagrame de sector
```

```python
import matplotlib.pyplot as plt

# ingresar el numero de varones y mujeres
while True:
    try:
        males = int(input("Ingrese el número de varones: "))
        females = int(input("Ingrese el número de mujeres: "))
        if males < 0 or females < 0:
            print("El número de varones y mujeres no puede ser negativo.")
        else:
            break
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número entero.")

total = males + females

if total == 0:
    print("No se ingresaron datos para varones y mujeres.")
else:
    # calcular el porcentaje
    percentage_males = (males / total) * 100
    percentage_females = (females / total) * 100

    print(f"Porcentaje de varones: {percentage_males:.2f}%")
    print(f"Porcentaje de mujeres: {percentage_females:.2f}%")

    #crear el diagrama de sector
    labels = ['Varones', 'Mujeres']
    sizes = [percentage_males, percentage_females]
    colors = ['skyblue', 'lightcoral']

    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    ax.axis('equal')
    ax.set_title('Distribución por Género')

    plt.show()
```
